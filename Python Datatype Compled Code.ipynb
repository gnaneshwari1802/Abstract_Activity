## 3-Creating variables and assigning values<a href="#3-Creating-variables-and-assigning-values" class="anchor-link">¶</a>

To create a variable in Python, all you need to do is specify the
variable name, and then assign a value to it.

= Python uses = to assign values to variables There's no need to declare
a variable in advance (or to assign a data type to it) Assigning a value
to a variable itself declares and initializes the variable with that
value. There's no way to declare a variable without assigning it an
initial value.
